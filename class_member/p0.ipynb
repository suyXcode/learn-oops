# ðŸ”¸ Class Members (Class Variables)

- Shared by all objects of the class.
- Declared inside the class but outside methods.
- Accessed using ClassName.variable_name.
  
---

```python
class Student:
    college = "KNIPSS"  # class variable (shared by all)

    def __init__(self, name, roll):
        self.name = name   # instance variable
        self.roll = roll

# Accessing class member
print(Student.college)  # KNIPSS

# Accessing through object
s1 = Student("Suyash", 101)
print(s1.college)  # KNIPSS
```

# ðŸ”¸ Object Members (Instance Variables)

- Belong to each individual object.
- Created inside the constructor (__init__) or methods using self.
- Accessed with object_name.variable_name.

---

```python
class Student:
    def __init__(self, name, roll):
        self.name = name       # object/instance variable
        self.roll = roll

s1 = Student("Suyash", 101)
s2 = Student("Aman", 102)

print(s1.name)  # Suyash
print(s2.name)  # Aman
```
