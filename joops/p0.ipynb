```python
class Solution:
    def intToRoman(self, num):
        values = [
            1000, 900, 500, 400,
            100, 90, 50, 40,
            10, 9, 5, 4, 1
        ]
        symbols = [
            "M", "CM", "D", "CD",
            "C", "XC", "L", "XL",
            "X", "IX", "V", "IV", "I"
        ]
        
        result = ""
        for i in range(len(values)):
            while num >= values[i]:
                result += symbols[i]
                num -= values[i]
        return result
```

```python
def int_to_roman(num):
    roman_dict = {1000: "M",900: "CM",500: "D",400: "CD",100: "C",90: "XC",50: "L",40: "XL",10: "X",9: "IX",5: "V",4: "IV",1: "I"}
    roman = ""
    for value in roman_dict:
        while num >= value:
            roman += roman_dict[value]
            num -= value
    return roman

# Test for 3999
print(int_to_roman(104))
```

# Static Method ğŸ‘
- It is used for utility function.
- It is neither related class nor object/instance.
- It is used to store extra functionality.
- It is used as helper/utility function inside a class.
- uses `@staticmethod` decorator and not conatoin any argument like cls,self to store class or object addresss passes another argument.
- `@staticmethod` decorator is used to define static method

**Syntax**

        class Class_name:
            @staticmethod
            def method_name():
                pass

        class.method_name()
        obj.method_name()

```python
class Bank:
    bname = "BOB"
    loc = "Noida"
    Ifsc = 'BOB123B'
    bmanager = 'Mr.Ashvik'

    def __init__(self, name, addr, sal):
        self.name = name
        self.addr = addr
        self.sal = sal

    @classmethod
    def banks_details(cls):
        print(f"Bank name     : {cls.bname}")
        print(f"Location      : {cls.loc}")
        print(f"IFSC code     : {cls.Ifsc}")
        print(f"Bank manager  : {cls.bmanager}")
    
    def customer_details(self):
        print(f"Customer Name : {self.name}")
        print(f"Address       : {self.addr}")
        print(f"Balance       : {self.sal}")

    def change_address(self):
        self.addr = input("Enter the new address of the customer : ")
        print("Address is changed successfully.")
    
    def deposit(self):
        amount = self.get_amount()
        self.sal += amount
        print(f"Amount {amount} deposited. New balance is {self.sal}.")
    
    def withdraw(self):
        amount = self.get_amount()
        if amount > self.sal:
            print("Insufficient balance.")
        else:
            self.sal -= amount
            print(f"Amount {amount} withdrawn. New balance is {self.sal}.")

    @staticmethod
    def get_amount():
        amount = int(input("Enter amount to deposit: "))
        return amount


c1 = Bank('Ashvik', 'Mumbai', 100000)
c2 = Bank('Abhinandan', 'Noida', 200000)



```

```python
class Addition:
    @staticmethod
    def add(x, y):
        return x + y
class Subtraction:
    @staticmethod
    def subtract(x, y):
        return x - y
class Product:
    @staticmethod
    def multiply(x, y):
        return x * y
class Division:
    @staticmethod
    def divide(x, y):
        if y != 0:
            return x / y
        else:
            return "Error! Division by zero."
class Calculator(Addition, Subtraction, Product, Division):
    pass

ob1 = Calculator()
print("Addition:", ob1.add(10, 5))
print("Subtraction:", ob1.subtract(10, 5))
print("Multiplication:", ob1.multiply(10, 5))
print("Division:", ob1.divide(10, 5))
```
