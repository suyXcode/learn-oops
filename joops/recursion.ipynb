```python
class Solution:
#using recursion
    def natnum(self, l, u):
        if l > u:
            return
        print(l)
        self.natnum(l + 1, u)
# using recursion
    def rnatnum(self,l,u):
        if l > u:
            return
        self.rnatnum(l + 1, u)
        print(l)
        
# using iterative method to reverse the string
    def reverse_s(self,s):
        rev=""
        for char in s:
            rev = char+rev
        return rev

```

```python
#creating object 
natural = Solution()
#call function using object
print("  Print  natural number :\n","=="*20)
natural.natnum(1, 5)
print("  Print reverse natural number :\n","=="*20)
#call function using object
natural.rnatnum(1, 5)
```

```python
#reverse the string
reverse = Solution()
print(reverse.reverse_s("Hello"))
```

```python
class Solve:
    def revs(self,s):
        if len(s)<=1:
            return s
        return self.revs(s[1:])+ s[0]

rev = Solve()
print(rev.revs("Hello"))
```

```python
class Solve:
    def factorial(self,n):
        if n==0 or n==1:
            return 1
        return n* self.factorial(n-1)

fact = Solve()
print(fact.factorial(6))
```

```python

class Solve:
    def sum_r(self,arr):
        if not arr:
            return 0
        return arr[0]+self.sum_r(arr[1:])
arr = Solve()
arr.sum_r([1,2,3,4,5,10])
```

```python
class Solve:
    def table(self,n,i=1):
        if i>10:
            return
        print(f"{n}x{i} = {n*1}")
        self.table(n,i+1)


tab = Solve()
tab.table(5)
```
