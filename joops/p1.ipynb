```python
# create a class which name is "School" property are school_name,address,principal, for three object 

# Creating class School
class School:
    sname = "BSS Jaunpur"
    address = "Jaunpur City"
    principal = "Ramesh"


# Creating 3 objects
Ankit = School()
Anuj = School()
Anhubhav = School()



#  Access properties using CLASS
print("-"*50)
print("Class Access:")
print("-"*50)
print("School Name :", School.sname)
print("School Location :", School.address)
print("School Principal :", School.principal)
print("-"*50)
print("\n")


#  Access properties using OBJECTS
print("-"*50)
print("Object Access:")
print("-"*50)

print("\nAnkit Details:")
print("-"*50)
print("School Name :", Ankit.sname)
print("Location :", Ankit.address)
print("Principal :", Ankit.principal)
print("-"*50)

print("\nAnuj Details:")
print("-"*50)
print("School Name :", Anuj.sname)
print("Location :", Anuj.address)
print("Principal :", Anuj.principal)
print("-"*50)

print("\nAnubhav Details:")
print("-"*50)
print("School Name :", Anhubhav.sname)
print("Location :", Anhubhav.address)
print("Principal :", Anhubhav.principal)
print("-"*50)



# -----------------------------------------------------------
#  MODIFICATION FOR EACH OBJECT
# -----------------------------------------------------------
print("\n\nUpdating Individual Object Details...\n")

Ankit.sname = "JNV Jaunpur"
Ankit.address = "Jaunpur"
Ankit.principal = "Mr. A"

Anuj.sname = "DAV Public School"
Anuj.address = "Lucknow"
Anuj.principal = "Mrs. B"

Anhubhav.sname = "St. Xavier School"
Anhubhav.address = "Varanasi"
Anhubhav.principal = "Mr. C"


# Show updated details
print("-"*50)
print("After Object Modifications:")
print("-"*50)

print("\nAnkit Updated Details:")
print("-"*50)
print("School Name :", Ankit.sname)
print("Location :", Ankit.address)
print("Principal :", Ankit.principal)

print("\nAnuj Updated Details:")
print("-"*50)
print("School Name :", Anuj.sname)
print("Location :", Anuj.address)
print("Principal :", Anuj.principal)

print("\nAnubhav Updated Details:")
print("-"*50)
print("School Name :", Anhubhav.sname)
print("Location :", Anhubhav.address)
print("Principal :", Anhubhav.principal)



# -----------------------------------------------------------
#  MODIFY CLASS PROPERTIES
# -----------------------------------------------------------
print("-"*50)
print("Updating CLASS Properties...")
print("-"*50)
School.sname = "Central School"
School.address = "New Delhi"
School.principal = "Dr. Sharma"


# Show class-level changes
print("-"*50)
print("After Class Modification (Class Access):")
print("-"*50)
print("School Name :", School.sname)
print("Location :", School.address)
print("Principal :", School.principal)


# Show object-level changes after class update
print("After Class Modification (Object Access):")

print("Ankit Details:")
print("-"*50)
print("School Name :", Ankit.sname)
print("Location :", Ankit.address)
print("Principal :", Ankit.principal)

print("Anuj Details:")
print("-"*50)
print("School Name :", Anuj.sname)
print("Location :", Anuj.address)
print("Principal :", Anuj.principal)

print("\nAnubhav Details:")
print("-"*50)
print("School Name :", Anhubhav.sname)
print("Location :", Anhubhav.address)
print("Principal :", Anhubhav.principal)
print("-"*50)
```
