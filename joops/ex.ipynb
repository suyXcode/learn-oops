```python
class Solve:
    def int_to_roman(self,num):
        roman_dict = {1000: "M",900: "CM",500: "D",400: "CD",100: "C",90: "XC",50: "L",40: "XL",10: "X",9: "IX",5: "V",4: "IV",1: "I"}
        roman = ""
        for value in roman_dict:
            while num >= value:
                roman += roman_dict[value]
                num -= value
        return roman
    
ro = Solve()
print(ro.int_to_roman(104))
```

```python
def int_to_binary(d,b=""):
    while d>0:
        re = d%2
        b=str(re)+b
        d//=2
    print(f"Binary is : {b}")

int_to_binary(10)

```

```python
def int_to_bin(d,b=""):
    if d==0:
        return "Binary is : "
    return  str(int_to_bin(d//2,b)) + str(d%2) 

print(int_to_bin(10))
```

```python
decimal = int(input("Enter a decimal number: "))
binary = ""

while decimal > 0:
    remainder = decimal % 2
    binary = str(remainder) + binary  
    decimal //= 2

print("Binary number:", binary)
```

```python
binary = int(input("Enter a binary number: "))
decimal = 0
base = 1   

while binary > 0:
    digit = binary % 10
    decimal += digit * base
    base *= 2
    binary //= 10

print("Decimal number:", decimal)
```

```python
def bin_to_int(bin,d=0,b=1):
    while bin>0:
        digit = bin % 10
        d+=digit * b
        b *=2
        bin //=10
    print(d)
bin_to_int(1010)
```

```python
a = "aaabcbbc"
i = 0
result = ""

while i < len(a):
    ch = a[i]
    count = 0
    while i < len(a) and a[i] == ch:
        count += 1
        i += 1
    result += ch + str(count)

print(result)
```
