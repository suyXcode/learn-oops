```python
class CarShowroom:
    # -------- CLASS VARIABLES --------
    brand = "BMW"
    head = "Suyash Singh"
    location = "Noida"
    est_year = 2005

    # -------- INSTANCE VARIABLES --------
    def __init__(self, model, color, price, year):
        self.model = model
        self.color = color
        self.price = price
        self.year = year

    # -------- INSTANCE METHOD --------
    def display(self):
        print("---- BMW Car Showroom Details ----")
        print("Showroom Head :", CarShowroom.head)
        print("Brand Name    :", CarShowroom.brand)
        print("Location      :", CarShowroom.location)
        print("Established   :", CarShowroom.est_year)
        print("----------------------------------")
        print("Car Model     :", self.model)
        print("Car Color     :", self.color)
        print("Car Price     :", self.price)
        print("Manufacture Y :", self.year)
        print("----------------------------------")

    # -------- CLASS METHOD to change showroom details --------
    @classmethod
    def update_showroom(cls, new_head, new_location):
        cls.head = new_head
        cls.location = new_location
# -------- CREATING OBJECTS --------
car1 = CarShowroom("X5", "Black", "80 Lakh", 2023)
car2 = CarShowroom("X3", "White", "60 Lakh", 2022)
# Displaying initial details
car1.display()
car2.display()
# Updating showroom details
CarShowroom.update_showroom("Ankit Sharma", "Gurgaon")
# Displaying updated details
car1.display()
car2.display()  

```
