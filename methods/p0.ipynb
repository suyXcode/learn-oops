# ‚öôÔ∏è 3Ô∏è‚É£ Types of Methods in a Class

    There are three main types of methods:

### üß† (i) Instance Method (Object Method)

- Operates on instance variables.
- First parameter is always self.
- Accessed by using object name.

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):  # instance method
        print(f"Name: {self.name}, Marks: {self.marks}")

s1 = Student("Suyash", 95)
s1.display()
```

# üè´ (ii) Class Method

- Works with class variables.
- Decorated with @classmethod.
- Takes cls (class reference) as the first argument.
- Accessed by class name or object.

```python
class Student:
    college = "KNIPSS"

    @classmethod
    def show_college(cls):
        print("College Name:", cls.college)

Student.show_college()  # via class
```

# üß∞ (iii) Static Method

- Does not depend on instance or class variables.
- Defined using @staticmethod.
- Can be used for utility work.

```python
class Math:
    @staticmethod
    def add(a, b):
        return a + b

print(Math.add(5, 3))  # Output: 8
```

# üß† Summary Table of Method Types
---


| Method Type  | Decorator       | First Parameter | Can Access                          | Called By      |
| ------------ | --------------- | --------------- | ----------------------------------- | -------------- |
| **Instance** | None            | `self`          | Instance variables                  | Object         |
| **Class**    | `@classmethod`  | `cls`           | Class variables                     | Class / Object |
| **Static**   | `@staticmethod` | None            | Neither class nor instance directly | Class / Object |
